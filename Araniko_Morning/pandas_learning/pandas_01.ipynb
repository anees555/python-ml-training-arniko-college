```python
dict2 = {
    "Students": ["student1", "student2", "student3", "student1", "student5", "student6", "student7"],

    "Roll no": ['005', '003', np.nan, '005', '001', '012', np.nan],

    "Gender": ["Male", "Female", "Male", "Male", np.nan, "Female", "Female"],

    "Marks_obtained": [65, 56, 78, 65, 49, np.nan, 34]
}
```

```python
dict2
```

```python
student_df = pd.DataFrame(dict2)
```

```python
student_df
```

### to check the null value

```python
student_df.isnull().sum()
```

#### to check the duplicate value in the record

```python
student_df.duplicated().sum()
```

### Remove the duplicate value

```python
student_df = student_df.drop_duplicates()
```

```python
student_df.dropna()
```

### Mean and mode imputation

```python
student_df
```

```python
gender_mode = student_df['Gender'].mode()[0]
```

```python
gender_mode
```

```python
marks_mean = student_df['Marks_obtained'].mean()
```

```python
marks_mean
```

```python
student_df['Gender'] = student_df['Gender'].fillna(gender_mode)
```

```python
student_df
```

### Filling marks

```python
student_df['Marks_obtained'] = student_df['Marks_obtained'].fillna(marks_mean)
```

```python
student_df
```

```python
student_df = student_df.dropna()
```
