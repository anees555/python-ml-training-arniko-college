## using mean in numpy array

```python
array1 = np.array([1, 5, 7, 8])
mean = np.mean(array1)
```

```python
print(mean)
```

```python
array2 = np.array([[6, 7, 8, 10],
                   [8, 3, 5, 12],
                   [10, 23, 4, 6],
                   [10, 12, 16, 18]])
```

```python
array2
```
