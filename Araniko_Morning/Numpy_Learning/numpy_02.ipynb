```python
import numpy as np
```

### Concatenation in numpy array

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([2, 2, 2 ,2])

```

```python
arr1, arr2
```

```python
arr3 = np.concatenate((arr1, arr2))
```

```python
arr3
```

```python
arr1, arr2
```

```python
arr4 = np.concatenate((arr1, arr2), axis=0)
```

```python
arr4
```

```python
arr1 = np.array([[1, 2, 3, 4]])
arr2 = np.array([[2, 2, 2 ,2]])
```

```python
arr3 = np.concatenate((arr1, arr2), axis = 0)
```

```python
arr3
```

```python
arr4 = np.concatenate((arr1, arr2), axis = 1)
arr4
```

```python
array5 = np.array([[4, 5, 6, 8],
                   [9, 9, 9, 9]])
array6 = np.array([[6, 6, 6, 6],
                   [1, 1, 1, 1]])
```

```python
np.concatenate((array5, array6), axis=1)
```

```python
np.concatenate((array5, array6), axis=0)
```

```python
np.vstack((array5, array6))
```

```python
np.hstack((array5, array6))
```

```python
array7 = np.array([[1, 2, 3],
                  [4, 5, 6]])
array8 = np.array([[2, 2, 2]])
```

```python
np.vstack((array7, array8))
```
