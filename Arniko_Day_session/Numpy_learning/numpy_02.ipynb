### Arithmetic operation in numpy

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
print(arr1 + arr2)
```

```python
arr3 = np.array([[4, 6, 7],
                [8, 4, 3],
                [9, 2, 1]])
arr4 = np.array([[6, 7, 7],
                [5, 4, 3],
                [3, 2, 1]])

arr3, arr4
```

```python
arr3 * arr4
```

## for matrix multiplication

```python
arr3 @ arr4
```
