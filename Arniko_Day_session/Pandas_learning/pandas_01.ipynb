```python
student_dict = {
    "Students": ["Student1", "Student2", "Student1", "Student4", "Student5", "Student6", "Student7", "Student8"],
    "Roll no": [1, np.nan, 1, 3, 4, 5, 6, 7],
    "Gender": ["Male", "Female", "Male", np.nan, "Male", "Female", "Female", "Male"],
    "Marks_obtained": [56, 76, 56, 43, 67, np.nan, 54, 75]

}
```

```python
student_df = pd.DataFrame(student_dict)
```

```python
student_df
```

### To check the duplicate value

```python
student_df.duplicated().sum()
```

```python
student_df = student_df.drop_duplicates()
```

```python
student_df
```

```python
student_df.dropna()
```

### Imputation method : (mean, median and mode)

```python
gender_mode = student_df["Gender"].mode()[0]
```

```python
gender_mode
```

```python
student_df["Gender"] = student_df["Gender"].fillna(gender_mode)
```

```python
student_df
```

```python
marks_mean = student_df["Marks_obtained"].mean()
```

```python
marks_mean = np.round(marks_mean, 2)
marks_mean
```

```python
student_df["Marks_obtained"] = student_df["Marks_obtained"].fillna(marks_mean)
```

```python
student_df
```

```python
student_df = student_df.dropna()
```

```python
student_df
```
